**5. Abre una imagen a color, imprime su tamaño, el número de píxeles y despliega la imagen con `matplotlib`**

```python
import cv2
import matplotlib.pyplot as plt
minion = cv2.imread('minion.jpg')
print(minion.shape)
print(minion.size)
minion=cv2.cvtColor(minion,cv2.COLOR_BGR2RGB)
plt.imshow(minion)
```

**7. Convierte la imagen anterior a escala de grises**

```python
minion_gray=cv2.cvtColor(minion,cv2.COLOR_RGB2GRAY)
plt.imshow(minion_gray, cmap='gray')
```

**8. Visita cada una de las posiciones de la imagen anterior (en escala de grises). Imprime las coordenadas de los píxeles que tengan un valor menor o igual a 120**

```python
minion_array= np.array(minion_gray)
coordenadas = []
for i in range(minion_array.shape[0]):
  for j in range(minion_array.shape[1]):
    if minion_array[i, j] <= 120:
      coordenadas.append((i, j))
print(coordenadas)
```

1.   Crea una imagen con dimensiones 600 x 600.
2.   Divide la imagen anterior en tres partes iguales y haz que se muestren en verde, blanco y rojo (respectivamente).
3.   Coloca un mensaje patrio como título de la imagen.


```python
image = np.zeros((600, 600, 3), dtype=np.uint8)
for i in range(3):
  color = [0, 255, 0] if i == 0 else [255, 255, 255] if i == 1 else [255, 0, 0]
  image[:, i * 200 : (i + 1) * 200, :] = color
plt.title("¡Viva Mexico!")
plt.imshow(image)
plt.show()
```
