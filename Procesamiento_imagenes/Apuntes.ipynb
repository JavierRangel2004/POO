## 6. Tipos de sensores

- **CMOS:** Un sensor CMOS (Complementary Metal-Oxide-Semiconductor) es un tipo de sensor de imagen utilizado en cámaras digitales, videocámaras, teléfonos inteligentes y una variedad de dispositivos electrónicos que requieren captura de imágenes. Este sensor es una parte fundamental en la conversión de la luz en señales eléctricas que luego se procesan para crear imágenes digitales. Esta tecnología permite la integración de circuitos de control y lógica directamente en el mismo chip del sensor, lo que los hace más versátiles y eficientes en cuanto a energía que sus contrapartes, los sensores CCD (Dispositivo de Acoplamiento de Carga).
- **CCD:** Un sensor CCD (dispositivo de carga acoplada, por sus siglas en inglés Charge-Coupled Device) es un tipo de sensor de imagen utilizado en cámaras digitales y otros dispositivos para capturar imágenes. La operación de un sensor CCD se basa en la conversión de la luz en cargas eléctricas. Cuando la luz incide en el sensor CCD, los fotones de luz liberan electrones en las celdas fotosensibles del sensor, generando una carga eléctrica proporcional a la cantidad de luz recibida. Luego, estas cargas se desplazan secuencialmente a través de las celdas del sensor, mediante un proceso llamado transferencia de carga, hacia una región de lectura. Allí, la cantidad de carga en cada celda se mide y se convierte en información digital que representa la imagen capturada.
- **Filtro de Bayer:** Esta matriz de filtros está compuesta por patrones repetitivos de filtros de color rojo, verde y azul (RGB), en una disposición específica. La matriz de Bayer suele tener una estructura de mosaico, donde aproximadamente la mitad de los píxeles son filtros verdes, alrededor de un cuarto son rojos y otro cuarto son azules. La razón detrás de esta disposición es que el ojo humano es más sensible al verde, por lo que los sensores capturan más información en verde para mejorar la percepción del color y la luminosidad en las imágenes.
Para obtener una imagen en color a partir de un sensor de imagen con filtro de Bayer, se utiliza un proceso llamado interpolación de color. Cada píxel del sensor solo captura información de un color (rojo, verde o azul), por lo que se utiliza un algoritmo para estimar los valores de color faltantes en cada píxel en función de los píxeles vecinos. Esto implica combinar la información de los píxeles adyacentes con diferentes filtros de color para generar una imagen completa en color.

## 7. Sistema Visual Humano

- **Introducción:** Convierte luz en imágenes a través del globo ocular. La retina contiene conos y bastones para interpretar ondas.
- **Visión:** Aproximadamente 576 MP de resolución.
  
  ### Funciones de Conos y Bastones
  
  **Conos:**
  - Detectan detalles finos.
  - Perciben colores: rojo, verde y azul.
  - Funcionan mejor con luz.
  
  **Bastones:**
  - Detectan formas generales.
  - Interpretan en escala de grises.
  - Funcionan en ausencia de luz.



```python
# Librerías
# import, as: permite importar librerías y asignarles un alias.
import numpy as np # pip install numpy
import math # pip install math
import cv2 as opencv # pip install opencv-python
import matplotlib.pyplot as plt # pip install matplotlib
from skimage import data # pip install scikit-image

# Cargando una imagen de ejemplo
image = data.camera()  # Esta es una imagen en escala de grises
imageRGB = data.astronaut()  # Esta es una imagen en RGB
```

```python
# Fundamentos de Python

# print(): imprime un mensaje o variable en consola.
print("Hello, World!")

# input(): recoge un valor ingresado por el usuario.
user_input = input("Enter your name: ")
print("Hello, " + user_input)

# len(): devuelve la longitud de un objeto.
length = len("Hello, World!")
print(f"length: {length}")

# int(), float(), str(): convierten valores a entero, flotante o cadena respectivamente.
integer_value = int("5")
float_value = float("5.2")
string_value = str(5)
print(f"integer_value: {type(integer_value)}, float_value: {type(float_value)}, string_value: {type(string_value)}")

# list(), tuple(): convierten valores a lista o tupla respectivamente.
list_value = list((1, 2, 3))
tuple_value = tuple([1, 2, 3])
print(f"list_value: {type(list_value)}, tuple_value: {type(tuple_value)}")

# range(): devuelve una secuencia de números.
numbers = range(5)
print(f"numbers: {numbers}")

# type(): devuelve el tipo de un objeto.
type_of_value = type(5)
print(f"type_of_value: {type_of_value}")

# sum(): devuelve la suma de todos los elementos de una lista o tupla.
total_sum = sum([1, 2, 3, 4, 5])
print(f"total_sum: {total_sum}")

# max(), min(): devuelven el máximo y mínimo valor de una lista o tupla respectivamente.
maximum_value = max([1, 2, 3, 4, 5])
minimum_value = min([1, 2, 3, 4, 5])
print(f"maximum_value: {maximum_value}, minimum_value: {minimum_value}")
```

```python
# OpenCV

# cvtColor(): convierte una imagen de un espacio de color a otro.
"""gray_image = opencv.cvtColor(image, opencv.COLOR_BGR2GRAY)"""
# cvtColor(): en este caso, dado que la imagen ya está en escala de grises, invertiremos los colores.
inverted_image = 255 - image

# imread(): lee una imagen.
image2 = opencv.imread("path_to_image.jpg")

# imwrite(): guarda una imagen. (Aunque la imagen es de skimage, OpenCV puede guardarla)
opencv.imwrite("inverted_image.jpg", inverted_image)

# split(): divide una imagen en sus canales de color.
red_channel, green_channel, blue_channel = opencv.split(imageRGB)
```

```python
# Numpy

# max(), min(), mean(): similar a las funciones nativas pero para arrays de numpy.
array = np.array([1, 2, 3, 4, 5])
max_value = np.max(array)
min_value = np.min(array)
mean_value = np.mean(array)
print (f"max_value: {max_value}, min_value: {min_value}, mean_value: {mean_value}")

# arange(): devuelve valores espaciados uniformemente dentro de un intervalo.
values = np.arange(0, 10, 2)
print (f"values: {values}")

# random.randint(): devuelve un entero aleatorio.
random_integer = np.random.randint(0, 10)
print (f"random_integer: {random_integer}")

# linspace(): devuelve números espaciados uniformemente durante un intervalo especificado.
linspace_values = np.linspace(0, 1, 5)
print (f"linspace_values: {linspace_values}")

# where(): devuelve índices donde una condición es verdadera.
indices = np.where(array > 3)
print (f"indices: {indices}")

# random.seed(): semilla para hacer reproducibles los resultados aleatorios.
np.random.seed(0)

# random.random(): devuelve un número flotante aleatorio entre 0 y 1.
random_float = np.random.random()
print (f"random_float: {random_float}")

# ones(): devuelve un nuevo array de forma y tipo dado, relleno de unos.
ones_array = np.ones((2, 2))
print (f"ones_array: {ones_array}")

# unique(): encuentra elementos únicos de un array.
unique_elements = np.unique([1, 2, 2, 3, 3, 3, 4, 4, 4, 4])
print (f"unique_elements: {unique_elements}")
```

```python
# Matplotlib

# pyplot.imshow(): muestra una imagen.
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.show()

plt.imshow(inverted_image , cmap='gray')
plt.title("Inverted Image")
plt.show()

plt.imshow(imageRGB)
plt.title("RGB Image")
plt.show()

# plt.imsave(): guarda una imagen usando Matplotlib.
plt.imsave("matplotlib_saved_image.jpg", image, cmap='gray')
```
